```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#RETRIEVING THE DATA
#Retriveing the Original Table as a CSV file from Google Sheets
df = pd.read_csv('AP Scores.csv')

df
```

```python
#ASSESSING DATA TYPE STATUS
#Assessing the Data Type Make Up/Count of Non-Null Values by Column
df.info()
```

```python
#CLEANING THE DATA
#Misspelled word in Column Head
df.rename(columns = {"Test Achievment": "Test Achievement"}, inplace = True)
df
```

```python
#CLEANING THE DATA
#Getting rid of the $'s and ,'s by making a new column (MHI_replace) 
#with the same numeric values as old "Median Household Income" column
# and making it of an integer value

df["MHI_replace"] = df["Median Household Income"].str.replace("$","").str.replace(",","")
df["MHI_replace"] = pd.to_numeric(df["MHI_replace"])
df["MHI_replace"]

df
```

```python
#CLEANING THE DATA
#Getting rid of the $'s and ,'s by making a new column (PCI_replace) 
#with the same numeric values as old "Per Capita Income" column
# and making it of an integer value

df["PCI_replace"] = df["Per Capita Income"].str.replace("$","").str.replace(",","")
df["PCI_replace"] = pd.to_numeric(df["PCI_replace"])
df["PCI_replace"]

df
```

```python
#CLEANING THE DATA
#Dropping the old "Median Household Income" column and renaming
#the new "MHI_replace" column as "Median Household Income"

df = df.drop(columns=['Median Household Income'])
df.rename(columns = {"MHI_replace": "Median Household Income"}, inplace = True)
df
```

```python
#CLEANING THE DATA
#Dropping the old "Per Capita Income" column and renaming
#the new "PCI_replace" column as "Per Capital Income"
df = df.drop(columns=['Per Capita Income'])
df.rename(columns = {"PCI_replace": "Per Capita Income"}, inplace = True)
df
```

```python
#CLEANING THE DATA
#Rearranging the columns as they originally were formatted
df = df[["County", "School", "Exams with Passing Scores", "Exams Taken", "Seniors Enrolled", "Test Achievement", "Test Participation", "Per Capita Income", "Median Household Income", "% below Poverty Line"]]
df
```

```python
#CLEANING THE DATA
#Deleting rows within the Test Achievment column that have null values "NaN"
df=df[df["Test Achievement"].notnull()]
df
```

```python
#ASSESSING DATA TYPE STATUS
#Reassessing the Data Type Make Up/Count of Stripped Data of Remaining Columns
df.info()
```

```python
#ANALYZING DATA
#Provides the count, mean, std, min, quartiles, median, and max values of Test Achievment Column
df.describe()
```

```python
#ANALYZING DATA
#Nassau County Correlation Table
nassau_county = df.loc[df.County == "Nassau"]

corr_mat_nassau_county = nassau_county.corr()
corr_mat_nassau_county
```

```python
#ANALYZING DATA
#Nassau County Correlation Matrix

plt.figure(figsize = (10,8))
plot = sns.heatmap(nassau_county.corr().round(2), annot = True)
plot.set_title("Nassau County: Analyzing Educational Inequity in the AP Program")
```

```python
#ANALYZING DATA
#Suffolk County Correlation Table

suffolk_county = df.loc[df.County == "Suffolk"]

corr_mat_suffolk_county = suffolk_county.corr()
corr_mat_suffolk_county
```

```python
#ANALYZING DATA
#Suffolk County Correlation Matrix

plt.figure(figsize = (10,8))
plot = sns.heatmap(suffolk_county.corr().round(2), annot = True)
plot.set_title("Suffolk County: Analyzing Educational Inequity in the AP Program")
```
